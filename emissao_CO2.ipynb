# Foi selecionado a coluna total de emissões que compõe todas as etapas para processamento do produto até a venda no varejo.

# Selecionado os 7 alimentos que mais emitem CO2, acima de 10kg de CO2 por kg de alimento;

```python
dados.query('Total_emissions>10').Total_emissions.sort_values(ascending=False)
```

```python
plt.figure(figsize=(20,10))
plt.title('Alimentos que apresentam maiores emissões de CO2 em kg', fontsize=20)

ax=sns.barplot(data=dados.query('Total_emissions>10'), x='Food product', y='Total_emissions', palette="Blues_d", orient='v')
ax =plt.ylabel('Total de Emissão de CO2 em kg', fontsize=18)
ax =plt.xlabel('Alimentos em kg', fontsize=18)
```

# Através do gráfico acima, foram selecionados os alimentos que emitem acima de 10kg de CO2 por kg de alimento, segue quem são eles: carne de gado, carne de carneiro, queijo, carne(gado leiteiro), chocolate preto, café, camarão(produzido em cativeiro), respectivamente.

```python
plt.figure(figsize=(20,10))
plt.title('Total de Emissão de CO2 x produto', fontsize=30)
sns.scatterplot(data=dados.query('Total_emissions>7'), x='Food product', y='Total_emissions', s=300, sizes=25 )
ax =plt.ylabel('Total de Emissão de CO2 em kg', fontsize=18)
ax =plt.xlabel('Alimentos em kg', fontsize=18)
```

## Como foram usados no experimento vários países, decidi importar as bases de dados dos países envolvidos:

#### Estudando os países: dados extraídos  https://ourworldindata.org/environmental-impacts-of-food

```python
paises = pd.read_csv('share-global-food-emissions.csv')
paises.head()
```

```python
paises.columns
```

```python
paises.Year.max()
```

```python
paises.Year.min()
```

```python
grupo = paises.groupby(by='Entity' , axis=0).mean()
grupo
```

# O Brasil de 1990 a 2015, apresentou uma participação global média de emissão de 9.72 % 

```python
grupo.iloc[26:27]['Share of global food emissions']
```

```python
grupo[['Share of global food emissions']].describe()
```

## Observa-se que 75% dos dados encontram-se com até 0.30% de participação mundial em emissão, na produção de alimentos. A China, USA são exemplos de países que já há muito tempo tem participação mundial na emissão de CO2, mas fica uma pergunta, quem são os países em industrialização e em desenvolvimento que compõem a maior parcela de emissão mundial? países que já foram citados devem estar na maior participação desses valores, mas quanto isso representa? 

```python
grupo[['Share of global food emissions']].max()
```

```python
grupo2 = grupo.rename(columns={'Share of global food emissions': "participacao"})
grupo2
```

```python
grupo2.query('participacao > 9')
```

## Para as mesmas faixas de ano a maior participação mundial em poluição foi atribuída ao desenvolvimento com 53.43% do total após vem a industrialização com 24.19%

```python
entidades = grupo2.query('participacao>3').participacao.sort_values(ascending=False)
entidades
```

```python
plt.figure(figsize=(20,8))

entidades.plot(grid=True)
plt.title('Entidades e países responsáveis pelas maiores emissões', fontsize=18)
```
